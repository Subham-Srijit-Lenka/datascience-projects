## Stock price analysis

```python
import pandas as pd
```

```python
df = pd.read_csv("A.csv", parse_dates=["Date"], index_col="Date")
print(df.head())
```

```python
import numpy as np
```

```python
print(df.info())
print(df.describe())
```

```python
df["Daily Return"] = df["Adj Close"].pct_change()
```

```python
print(df.head())
```

```python
df['MA20'] = df['Adj Close'].rolling(window=20).mean()
df['MA50'] = df['Adj Close'].rolling(window=50).mean()

print(df[['Adj Close', 'MA20', 'MA50']].head(60))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,6))
plt.plot(df["Adj Close"], label="Adj Close", alpha=0.5)
plt.plot(df["MA20"], label="20 day MA", linewidth=2)
plt.plot(df["MA50"], label="50 day MA", linewidth=2)
plt.legend()
plt.show()
```
